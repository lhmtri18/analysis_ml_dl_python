```python
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
reset_graph()


import numpy as np
from sklearn.datasets import fetch_california_housing
housing= fetch_california_housing()
m,n= housing.data.shape#shape (20640,8)
#print(housing.data.shape)
housing_data_plus_bias= np.c_[np.ones((m,1)),housing.data]
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler().fit(housing.data)
scaled_housing_data= scaler.transform(housing.data)
scaled_housing_data_plus_bias= np.c_[np.ones((m,1)),scaled_housing_data]


#model linear regression, complete

learning_rate = 0.01
#X= tf.constant(scaled_housing_data_plus_bias,dtype=tf.float32, name='X')
#y= tf.constant(housing.target.reshape(-1,1),dtype=tf.float32, name='y')#reshape (-1,1), stay unchange row, and column is 1

X = tf.placeholder(tf.float32, shape=[100, 9], name="X")
y = tf.placeholder(tf.float32, shape=[100, 1], name="y")

theta = tf.Variable(tf.random_uniform([n+1,1],-1.0,1.0,seed=42),name='theta')#shape is [n+1,1], value of range is -1 to 1
y_pred= tf.matmul(X,theta,name='predictions')
error= y_pred-y
mse= tf.reduce_mean(tf.square(error),name='mse')
#gradients = tf.gradients(mse, [theta])[0] #compute automatic gradient
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(mse)

init= tf.global_variables_initializer()
n_epochs = 10
batch_size=100
n_batches= int(np.ceil(m/batch_size))#207,m= 20640

def fetch_batch(epoch, batch_index, batch_size):
  np.random.seed(epoch * n_batches + batch_index) 
  indices = np.random.randint(m, size= batch_size)#index t or f
  X_batch= scaled_housing_data_plus_bias[indices]
  y_batch= housing.target.reshape(-1,1)[indices]
  return X_batch, y_batch


with tf.Session() as sess:
  sess.run(init)
  for epoch in range(n_epochs):
    for batch_index in range(n_batches):
      X_batch, y_batch = fetch_batch(epoch, batch_index, batch_size)
      sess.run(training_op, feed_dict={X: X_batch, y: y_batch})  
      if epoch % 100 == 0:
        print("Epoch", epoch, "MSE =", mse.eval())   
    best_theta = theta.eval()
```
